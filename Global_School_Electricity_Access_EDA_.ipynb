## Electricity Access Trends Over Time

### Subtask:
Investigating how the average 'Electricity_Access_Percent' has changed over the years using a Plotly line chart.


**Reasoning**:
To analyze the trend of electricity access over time, I will group the DataFrame by 'Year', calculate the mean of 'Electricity_Access_Percent' for each year, and then use Plotly to visualize this trend with a line chart.



```python
import plotly.express as px

# Calculate the average 'Electricity_Access_Percent' for each 'Year'
avg_electricity_access_per_year = df.groupby('Year')['Electricity_Access_Percent'].mean().reset_index()

# Create a Plotly line chart
fig_line = px.line(
    avg_electricity_access_per_year,
    x='Year',
    y='Electricity_Access_Percent',
    title='Average Electricity Access Percentage Over Time',
    labels={'Electricity_Access_Percent': 'Average Electricity Access (%)'}
)

# Display the plot
fig_line.show()
```

The line chart visualizing the 'Average Electricity Access Percentage Over Time' provides the following insights:

**1. Early Period (2000-2008):**
*   From 2000 to 2008, the average electricity access percentage remained constant at 100%. This indicates that for the countries and years included in the dataset during this period, schools had full electricity access on average. It's important to note that this might reflect data availability rather than a universal 100% access globally; it could mean only countries with 100% access were reported or included.

**2. Initial Decline (2009-2011):**
*   There's a noticeable dip in the average access percentage starting around 2009, continuing through 2011. The average drops from 100% to approximately 98% in 2009, then to around 95-96% by 2011. This could be due to more comprehensive data becoming available for countries with lower access, or it could represent actual challenges in maintaining or expanding access during these years.

**3. Recovery and Stabilization (2012-2020):**
*   Following the dip, the average electricity access percentage shows a gradual recovery and stabilization. From 2012 onwards, the average access percentage steadily increases, reaching back towards the high 90s. By 2018-2020, the average is again very high, suggesting sustained efforts in improving electricity access in schools. The line becomes relatively flat in the later years, indicating a near-universal high level of average access in the reported data.

**Overall Trend:**
*   The overall trend suggests a highly successful electrification effort for schools, with the average access consistently being very high (mostly above 95%). The dip in the early 2010s might represent a period of broader data collection, including regions with varying access levels, before a concerted global effort brought the average back up. The trend is generally positive, moving towards increasing and sustaining high levels of electricity access for schools over the two decades covered.

## Top and Bottom Countries by Electricity Access

### Subtask:
Identifying and visualizing the top 10 and bottom 10 countries based on their average 'Electricity_Access_Percent' using Plotly bar charts.

**Reasoning**:
To identify the top and bottom 10 countries by electricity access, I will first calculate the average 'Electricity_Access_Percent' for each country, then sort these averages to get the top 10 and bottom 10, and finally visualize them using Plotly bar charts as requested.



```python
import plotly.express as px

# Calculate the average 'Electricity_Access_Percent' for each country
avg_electricity_access_per_country = df.groupby('Country_Name')['Electricity_Access_Percent'].mean().reset_index()

# Sort for top 10 countries
top_10_countries = avg_electricity_access_per_country.sort_values(by='Electricity_Access_Percent', ascending=False).head(10)

# Sort for bottom 10 countries
bottom_10_countries = avg_electricity_access_per_country.sort_values(by='Electricity_Access_Percent', ascending=True).head(10)

# Create a Plotly bar chart for the top 10 countries
fig_top10 = px.bar(
    top_10_countries,
    x='Country_Name',
    y='Electricity_Access_Percent',
    title='Top 10 Countries by Average Electricity Access',
    labels={'Electricity_Access_Percent': 'Average Electricity Access (%)'},
    color='Electricity_Access_Percent',
    color_continuous_scale=px.colors.sequential.Viridis
)
fig_top10.show()

# Create a Plotly bar chart for the bottom 10 countries
fig_bottom10 = px.bar(
    bottom_10_countries,
    x='Country_Name',
    y='Electricity_Access_Percent',
    title='Bottom 10 Countries by Average Electricity Access',
    labels={'Electricity_Access_Percent': 'Average Electricity Access (%)'},
    color='Electricity_Access_Percent',
    color_continuous_scale=px.colors.sequential.Plasma
)
fig_bottom10.show()
```

The bar charts provide insights into countries with the highest and lowest average electricity access percentages:

**1. Top 10 Countries by Average Electricity Access (`fig_top10`):**
*   This chart clearly shows that many countries have consistently maintained a 100% average electricity access for schools across the years included in the dataset. This indicates highly successful or fully implemented electrification programs in these regions.
*   Countries like Andorra, Armenia, Antigua and Barbuda, Aruba, Bahamas, Azerbaijan, Australia, and Barbados are among those demonstrating full average access.
*   The uniform height of the bars at 100% for most of these countries highlights a strong global achievement in providing school electricity access in certain regions.

**2. Bottom 10 Countries by Average Electricity Access (`fig_bottom10`):**
*   This chart reveals the countries facing significant challenges in providing electricity access to schools.
*   The average electricity access percentages in these countries are strikingly low, with some even falling below 5% (e.g., Chad, Central African Republic, Niger).
*   Countries like Chad, Central African Republic, Niger, Nepal, Sierra Leone, Burundi, Madagascar, and Guinea are consistently at the lower end of the spectrum, indicating persistent disparities and areas requiring urgent intervention.
*   The varying heights of the bars for these countries suggest different levels of struggle, but all are far from achieving universal access.

**Overall Insights:**

The visualizations underscore a stark contrast in electricity access levels globally. While many countries have achieved full or near-full electricity access for their schools, a significant number of countries, predominantly in sub-Saharan Africa and parts of Asia, are lagging far behind. This highlights the need for targeted interventions and support to address these disparities and improve educational infrastructure in underserved regions.

## Country-Specific Trends in Electricity Access

### Subtask:
Visualiziong the trend of 'Electricity_Access_Percent' over time for a selection of countries using Plotly line charts, possibly showing multiple countries on one plot for comparison. This will help identify country-specific progress or stagnation.


**Reasoning**:
To visualize country-specific trends, I will first select a diverse set of countries that represent both high and low electricity access. Then, I will filter the main DataFrame to include only these selected countries. Finally, I will use `plotly.express.line` to create a multi-line chart, allowing for easy comparison of 'Electricity_Access_Percent' over 'Year' for each chosen country.

**Reasoning**:
To visualize country-specific electricity access trends over time, I will select a sample of countries, filter the main DataFrame, and then generate a Plotly line chart showing 'Electricity_Access_Percent' against 'Year' for each selected country.



```python
import plotly.express as px

# 1. Select a few interesting countries for detailed analysis
selected_countries = ['Chad', 'Central African Republic', 'Niger', 'Andorra', 'Australia', 'Albania']

# 2. Filter the main DataFrame `df` to include only the data for these selected countries
df_selected_countries = df[df['Country_Name'].isin(selected_countries)]

# 3. Create a Plotly line chart
fig_country_trends = px.line(
    df_selected_countries,
    x='Year',
    y='Electricity_Access_Percent',
    color='Country_Name',
    title='Electricity Access Trends Over Time for Selected Countries',
    labels={
        'Electricity_Access_Percent': 'Electricity Access (%)',
        'Year': 'Year',
        'Country_Name': 'Country'
    }
)

# 7. Display the plot
fig_country_trends.show()
```

The line chart showing electricity access trends for selected countries provides several key insights:

**1. High Access Countries (Andorra, Australia, Albania):**
*   **Andorra and Australia** consistently show 100% electricity access throughout the years. This indicates a fully electrified school system in these countries, with no reported fluctuations or challenges in maintaining access. They serve as examples of universal access.
*   **Albania** also maintains a very high electricity access, hovering around the high 90s, with some slight variations but generally stable and successful access levels.

**2. Low Access Countries (Chad, Central African Republic, Niger):**
*   **Chad, Central African Republic, and Niger** consistently show very low electricity access percentages, often below 10% for most of the years. This highlights significant and persistent challenges in providing basic electricity to schools in these regions.
*   The trends for these countries appear relatively flat or show only marginal improvements over the years, indicating a lack of substantial progress in electrification efforts for schools during the observed period.
*   There are no clear upward trends, suggesting that these countries are still far from achieving widespread electricity access for their educational institutions.

**Overall Insights:**

The visualization starkly illustrates the global disparity in school electricity access. While some countries have achieved and maintained universal access, a critical group of nations, particularly in sub-Saharan Africa, faces severe and ongoing challenges. The flat or low-progress lines for these low-access countries underscore the urgent need for targeted interventions, investment, and sustainable solutions to improve educational infrastructure and opportunities in these regions. The data suggests that while overall average access might be high due to many countries having 100% access, the problem of energy poverty in schools remains acute in specific areas.
