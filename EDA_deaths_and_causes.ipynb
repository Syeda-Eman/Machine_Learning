# Exploratory Data Analysis (EDA)
## Deaths and Causes Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('deaths_and_causes_synthetic.csv')
df.head()
```

```python
# Summary statistics
df.describe()
```

```python
# Deaths by cause
df.groupby('Cause_of_Death')['Number_of_Deaths'].sum().sort_values().plot(kind='barh')
plt.title('Total Deaths by Cause')
plt.show()
```

```python
# Death trend over years
df.groupby('Year')['Number_of_Deaths'].sum().plot()
plt.title('Total Deaths Over Time')
plt.show()
```
